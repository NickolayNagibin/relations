```python
class otn():
  def __init__ (self, f, A, B):
    import numpy as np
    self.A = A
    self.B = B
    if type(f) != type(np.eye(2)):
      matrix = []
      for x in A:
        matrix.append([])
        for y in B:
          matrix[-1].append(int(f(x, y)))
      self.np = np.array(matrix)
    else:
      self.np = f

  def __add__ (self, other):
    q = (self.np+other.np).astype(bool).astype(int)
    return otn(q, self.A, self.B)

  def __abs__ (self):
    return np.sum(self.np)

  def __eq__ (self, other):
    return (self.A == other.A) and (self.B==other.B) and (self.np==other.np).all()

  def __mul__ (self, other):
    return otn(np.multiply(self.np, other.np), self.A, self.B)

  def dot(self, other):
    return otn(np.dot(self.np, other.np).astype(bool).astype(int), self.A, other.B)

  def obrat(self):
    return otn(np.transpose(self.np), self.B, self.A)

  def dom(self):
    res = np.dot(self.np, np.ones((len(self.B), 1))).astype(bool)
    r = []
    for q in range(len(self.A)):
      if res[q]:
        r.append(self.A[q])
    return r

  def rang(self):
    return self.obrat().dom()

  def ker(self):
    return self.dot(self.obrat())

  def __str__ (self):
    r = "Матрица: \n"
    r += str(self.np)
    r+="\nDom = "+ str(self.dom())+ "; Range = "+ str(self.rang())+"\n\n"
    return r

  def refl(self):
    return self + otn(np.eye(len(self.A)), self.A, self.A)

  def symm(self):
    return self + self.obrat()

  def trans(self):
    Q = otn(np.eye(len(self.A)), self.A, self.A)
    step = otn(self.np, self.A, self.A)
    sum = step
    for n in range(2, len(self.A)+1):
      step = step.dot(self)
      sum += step
    return sum









```

```python
import math
import numpy as np
A = [1,2,3,4,5,6,7]
B = [3,4,5,7,8,10]
def fp (x, y):
  return math.cos(38*x*x*y)>0.5

def fq(x, y):
  return math.sin(38*x*math.sqrt(y))<-0.5

Q = otn(fq, A, B)
P = otn(fp, A, A)
print("P")
print(P)
print("Q")
print(Q)
print("PoQ")
print(P.dot(Q))
print("Ker(Q)")
print(Q.ker())
print("[P]r")
print(P.refl())
print("[P]s")
print(P.symm())
print("[P]t")
print(P.trans())
print("[P]rs")
print(P.refl().symm())
print("[P]sr")
print(P.symm().refl())
print("[P]rs = [P]sr: ", P.refl().symm()==P.symm().refl())
print("[P]rt")
print(P.refl().trans())
print("[P]tr")
print(P.trans().refl())
print("[P]rt = [P]tr: ", P.refl().trans()==P.trans().refl())
print("[P]st")
print(P.symm().trans())
print("[P]ts")
print(P.trans().symm())
print("[P]st = [P]ts: ", P.symm().trans()==P.trans().symm())
print("[P]eq")
print(P.refl().symm().trans())
print(P.refl().trans().symm())

```
